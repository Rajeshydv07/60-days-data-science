# Day 6: Data Cleaning & Preprocessing
In this notebook, we will clean our messy dataset by handling missing values, removing duplicates, fixing data types, and creating new features.

```python
import pandas as pd
import numpy as np

# Load the messy dataset
df = pd.read_csv('messy_dataset.csv')
df.head()
```

### 1. Handling Missing Values
Let's check for missing values and handle them.

```python
# Check missing values
df.isnull().sum()
```

```python
# We can fill missing numerical values with median and categorical with mode.
# For 'Sales', since it is loaded as object due to nan or strings, we will handle its type first in the next section.
# Let's fill 'Category' missing values with mode
df['Category'] = df['Category'].fillna(df['Category'].mode()[0])

# Fill 'Profit' missing values with median
df['Profit'] = df['Profit'].fillna(df['Profit'].median())

df.isnull().sum()
```

### 2. Removing Duplicates
Check and drop duplicate rows.

```python
# Check duplicates
print('Duplicates:', df.duplicated().sum())

# Drop duplicates
df = df.drop_duplicates()
print('Duplicates after dropping:', df.duplicated().sum())
```

### 3. Fixing Data Types
Convert columns to appropriate data types.

```python
df.info()
```

```python
# Convert 'Sales' and 'Quantity' to numeric
df['Sales'] = pd.to_numeric(df['Sales'])
df['Quantity'] = pd.to_numeric(df['Quantity'])

# Convert date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df.info()
```

### 4. Creating New Features
We will create 2 new features:
1. **Processing Time**: Ship Date - Order Date
2. **Sales per Item**: Sales / Quantity

```python
# Feature 1: Processing Time (in days)
df['Processing Time'] = (df['Ship Date'] - df['Order Date']).dt.days

# Feature 2: Sales per Item
df['Sales per Item'] = df['Sales'] / df['Quantity']

df.head()
```

### 5. Saving the Cleaned Dataset

```python
df.to_csv('cleaned_dataset.csv', index=False)
print('Cleaned dataset saved successfully!')
```
