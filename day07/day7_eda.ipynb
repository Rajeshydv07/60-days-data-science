# Day 7: Exploratory Data Analysis (EDA)
Understanding patterns and outliers before modeling.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_theme(style='whitegrid')
```

### 1. Load Data

```python
df = pd.read_csv('day7_dataset.csv')
df.head()
```

```python
df.describe()
```

### 2. Plot Distributions
Visualizing distributions to understand the spread and central tendency of our data.

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df['Income'], bins=20, kde=True)
plt.title('Income Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df['Spending_Score'], bins=20, kde=True)
plt.title('Spending Score Distribution')

plt.tight_layout()
plt.show()
```

### 3. Identify Patterns and Outliers
Using boxplots to spot outliers and scatter plots for relationships.

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(y=df['Age'])
plt.title('Age Boxplot')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['Income'])
plt.title('Income Boxplot')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['Spending_Score'])
plt.title('Spending Score Boxplot')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Income', y='Spending_Score', hue='Category')
plt.title('Income vs Spending Score by Category')
plt.show()
```

### 4. Top 5 Insights

1. **Age Distribution:** The age of customers is normally distributed, primarily centered around 35 years old, though there is one extreme outlier at 110 years old that requires cleaning.
2. **Income Outliers:** Income is clustered around $60,000, but boxplots clearly show a few high-income outliers (>$200,000) that could skew mean-based models.
3. **Spending Score Range:** The spending score has a wide spread, mostly between 30 and 70, with at least one anomalous score exceeding 140.
4. **Category Popularity:** The scatter plot indicates an even distribution of customer preferences across Electronics, Clothing, Groceries, and Home categories.
5. **Lack of Strong Linear Correlation:** There is no immediate visual linear correlation between Income and Spending Score, suggesting customer spending is driven by factors other than just their income level.
